```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Discrete Time Square Wave
t = np.linspace(0, 1, 50, endpoint=False)
f = signal.square(2 * np.pi * 5 * t)
n = len(f)
plt.plot(t, f, "-o")
plt.ylim(-2, 2)

plt.show()
plt.grid()
plt.xlabel('Time')
plt.ylabel('Y')
plt.title('Square Wave Function')
```

```python
# Discrete time square wave in frequency
Fs = 50
fTrans = np.fft.fft(f)/n 
fTrans = fTrans[range(n/2)]
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(n/2)] # one side frequency range
print(len(fTrans))

plt.plot(frq,abs(fTrans),'r') # plotting the spectrum
plt.xlabel('Freq (Hz)')
plt.ylabel('|Y(freq)|')
plt.title('Frequency Amplitudes')

plt.show()
plt.grid()
#f, t, Sxx = signal.spectrogram(f)
```

```python
Pxx, freqs, bins, im  = plt.specgram(f)
plt.show()
print(freqs)
print(Pxx)
```
