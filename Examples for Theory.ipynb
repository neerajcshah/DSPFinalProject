```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Discrete Time Square Wave
t = np.linspace(0, 1, 50, endpoint=False)
f = signal.square(2 * np.pi * 5 * t)
plt.plot(t, f, "-o")
plt.ylim(-2, 2)

plt.show()
plt.grid()
```

```python
# Discrete time square wave in frequency
f, t, Sxx = signal.spectrogram(f)
```
